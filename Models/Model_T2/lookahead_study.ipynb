## population = 100, workers = 100, iter=5

## Posterior and run time

```python
gp = os.getcwd() + "/data/scripts_la_100p_100w/data_la"

pyabc.settings.set_figure_params("pyabc")

sampler_ids = ["STAT", "DYN_Wait", "DYN", "LA_fix", "LA"]
sampler_labels = ["STAT", "DYN_Wait", "DYN", "LA", "LA_fix"]

sampler_ids = ["STAT", "DYN", "LA_fix", "LA"]
sampler_labels = ["STAT", "DYN", "LA Pre", "LA Curr"]

colors = ["C1", "C2", "C3", "C4", "C0"]

data_dir = "data_la"
model_id = "bimodal"

pop_size = 100
worker_size = 144
sigma = 0.5
iters = 5

times = []
for sampler_id in sampler_ids:
    times_for_sampler = []
    for ix in range(iters):
        h = pyabc.History(
            f"sqlite:///{gp}/{model_id}__{sigma}__{sampler_id}__{pop_size}__{ix}.db", create=False)
        abc = h.get_abc()
        times_for_sampler.append(
            (abc.end_time - abc.start_time).total_seconds())
    times.append(times_for_sampler)


times = np.array(times)
times /= np.max(np.mean(times, axis=1))
print(np.mean(times, axis=1))
print(np.std(times, axis=1))
fig, axes = plt.subplots(ncols=2, figsize=(14, 4))

axes[0].bar(
    x=sampler_labels,
    height=np.mean(times, axis=1),
    yerr=np.std(times, axis=1),
    color=samplers_color
)
axes[0].set_title(f"Simulation time - {pop_size}p - {worker_size}w")

iters = 5
for sampler_ix, sampler_id in enumerate(sampler_ids):
    for ix in range(iters):
        h = pyabc.History(
            f"sqlite:///{gp}/{model_id}__{sigma}__{sampler_id}__{pop_size}__{ix}.db", create=False)
        pyabc.visualization.plot_kde_1d_highlevel(
            h, x="p", xname=r"$\theta_1$", xmin=-2, xmax=2,
            numx=200, color=samplers_color[sampler_ix],
            ax=axes[1],
            label=sampler_labels[sampler_ix] if ix == 0 else None,
        )
axes[1].set_ylim([0, 0.6])

plt.title(f"Posterior - {pop_size}p x {worker_size}w")
axes[1].legend()

plt.savefig(f"T2_time_posterior_{pop_size}px{worker_size}w.pdf", format="pdf", bbox_inches="tight")

```
